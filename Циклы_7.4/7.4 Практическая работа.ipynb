1. Напишите цикл, который выводит на печать все високосные годы начиная с 2020 года, но не позже 2050 года.

```python
# начнем с 2020 года
year = 2020

# цикл для каждого года начиная с 2020 до 2050 года
while year <= 2050:
    # проверим, является ли год високосным
    if year % 4 == 0:
        # если год делится на 100, то он должен быть делимым на 400
        if year % 100 == 0:
            if year % 400 == 0:
                print(year)
        else:
            print(year)
    year += 1
```

2. Студент Джонсон поступил в университет и взял в кредит 120 тысяч долларов под 6% годовых. Все шесть лет учёбы в бакалавриате и магистратуре он не сможет выплачивать кредит, и процент будет прибавляться к основному кредиту. Напишите программу, которая посчитает, сколько студент Джонсон будет должен банку к концу своего обучения.

```python
# начальный долг студента
principal = 120000

# процентные ставки (в десятичных долях)
rates = [0.06] * 6

# матрица выплат
payments = []

# для каждого года обучения
for i in range(6):
    # каждый год увеличивает долг на процентную ставку от текущего долга
    interest = principal * rates[i]
    total_payment = principal + interest
    
    # заносим платеж в матрицу
    payments.append(total_payment)
    
    # обновляем текущий долг
    principal = total_payment

# итоговое значение долга студента
final_debt = principal

print(f"Студент Джонсон будет должен банку {final_debt} долларов к концу своего обучения.")
```

Вспомним задачу о [зёрнах на шахматной доске](https://ru.wikipedia.org/wiki/%D0%97%D0%B0%D0%B4%D0%B0%D1%87%D0%B0_%D0%BE_%D0%B7%D1%91%D1%80%D0%BD%D0%B0%D1%85_%D0%BD%D0%B0_%D1%88%D0%B0%D1%85%D0%BC%D0%B0%D1%82%D0%BD%D0%BE%D0%B9_%D0%B4%D0%BE%D1%81%D0%BA%D0%B5). На первую клетку кладётся одно зерно, на вторую — два зерна, на третью — четыре, затем — восемь. С каждой клеткой количество зёрен увеличивается в два раза.

Напишите цикл, который выводил бы на печать порядковый номер клетки и количество зёрен на ней до тех пор, пока их количество на одной клетке не превысит 100 тысяч.


```python
def zernike(limit):
    grains = [1, 2, 4, 8]
    k = 0

    while grains[k] <= limit:
        print(f"На клетке {k+1} лежит {grains[k]} зерна.")
        k += 1
        if k < len(grains):
            grains[k] = grains[k-1] * 2
        else:
            grains.append(grains[-1] * 2)

    return grains


k = 0
limit = 100000
while zernike(limit)[k] <= limit:
    k += 1

print(f"\n\nТребуется {k} клеток, чтобы количество зёрен превысило 100 тысяч.")
```

4. Напишите программу, которая проверяла бы все числа от 0 до 30. Если число делится на два, его нужно напечатать. Но, если число делится и на четыре, его нужно пропустить.

```python
# Программа, которая проверяет все числа от 0 до 30
for i in range(31):
    # Проверяем, является ли число двукратным
    if i % 2 == 0:
        # Проверяем, является ли число четырекратным
        if i % 4 != 0:
            # Печатаем число, если оно не является четырекратным
            print(i)
```

  * Если число делится на семь, нужно вывести на печать «дурак совсем» и прервать выполнение программы. 
  * Во всех остальных случаях программа должна снова запросить число на ввод.
